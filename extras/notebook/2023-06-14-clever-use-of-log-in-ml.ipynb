# Logarithm function plot

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.001, 10, 1000)

y = np.log(x)

plt.plot(x, y, label='y = log(x)')
plt.xlabel('x')
plt.ylabel('log(x)')
plt.title('Logarithm Function')
plt.xlim(-1, 10)
plt.ylim(-5, 4)
plt.xticks(np.arange(-1, 11, 1))
plt.yticks(np.arange(-5, 5, 1))
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend()

plt.axvline(x=0, color='black')
plt.axhline(y=0, color='black')

plt.show()
```

# Bar chart visualization

```python
import matplotlib.pyplot as plt

x = ["A", "B", "C", "D"]
y = [5000, 20, 15, 30]

fig, ax = plt.subplots()
ax.bar(x = x, height = y)
plt.show()

fig, ax = plt.subplots()
ax.bar(x = x, height = y, log = True)
plt.show()
```

# Right skewed distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skewnorm

skewness = 10  # positive value for right skewed distribution
mean = 0
standard_deviation = 1

data = skewnorm.rvs(skewness, loc=mean, scale=standard_deviation, size=1000)

plt.hist(data, bins=50, density=True)
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('Right-Skewed Distribution')
plt.show()

# Note: adding +1 to prevent the logarithm from converting 
# zero values in data into negative infinity
data_with_log_transformation = np.log10(data + 1)

plt.hist(data_with_log_transformation, bins=50, density=True)
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('Right-Skewed Distribution - after logarithm transformation')
plt.show()
```
